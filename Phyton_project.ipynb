```python
import statistics

# Sample Data
ages = [30, 45, 22, 35, 50]
stress_levels = [5, 8, 3, 7, 6]  # Scale 1-10

# Calculate Means
mean_age = statistics.mean(ages)
mean_stress = statistics.mean(stress_levels)

print(f"Average Age: {mean_age}")
print(f"Average Stress Level: {mean_stress}")
```

```python
import numpy as np

# Sample Data: [Age, StressLevel (1-10)]
employees = np.array([
    [25, 3],
    [30, 7],
    [35, 5],
    [40, 8],
    [22, 2]
])

# 1. Calculate average of each column (Age and StressLevel)
# Axis 0 = column-wise mean
avg_age, avg_stress = np.mean(employees, axis=0)

print(f"Average Age: {avg_age}")
print(f"Average Stress Level: {avg_stress}")

# Alternative: If data is in separate arrays
# ages = np.array([25, 30, 35, 40, 22])
# print(np.mean(ages))
```

```python
import numpy as np

# Create a NumPy array of sleep hours (in hours)
# Example data: [7.5, 6.0, 8.2, 5.5, 7.0, 6.5]
sleep_hours = np.array([7.5, 6.0, 8.2, 5.5, 7.0, 6.5])

# Method 1: Use the np.min() function
min_hours_func = np.min(sleep_hours)

# Method 2: Use the .min() method on the array
min_hours_method = sleep_hours.min()

# Print the minimum sleep hours
print(f"The minimum sleep hours using np.min() is: {min_hours_func} hours")
print(f"The minimum sleep hours using .min() method is: {min_hours_method} hours")
```

```python
import pandas as pd

# Create sample DataFrames (df1 and df2)
df1 = pd.DataFrame({
    'Key': ['A', 'B', 'C'],
    'Value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'Key': ['B', 'C', 'D'],
    'Value2': [4, 5, 6]
})

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Perform an outer merge
merged_outer_df = pd.merge(df1, df2, how='outer', on='Key')

print("\nOuter Merged DataFrame:")
print(merged_outer_df)
```

```python
import pandas as pd

# Create the first DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

# Create the second DataFrame with some overlapping IDs
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [25, 30, 22, 35]
})

# Perform an inner join on the 'ID' column
# 'how="inner"' is optional as it is the default
merged_inner = pd.merge(df1, df2, on='ID', how='inner')

# Print the result
print(merged_inner)
```
